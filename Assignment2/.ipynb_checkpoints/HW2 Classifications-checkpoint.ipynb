```python
import csv
import matplotlib.pyplot as plt
import matplotlib.colors
import numpy as np
import pandas as pd
from sklearn import naive_bayes
from sklearn import svm
from sklearn.cross_validation import train_test_split, cross_val_score 
from sklearn import cluster
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sas7bdat import SAS7BDAT
%pylab inline
```

```python
with SAS7BDAT('chs2014_public.sas7bdat') as f:
    df = f.to_data_frame()
```

```python
df.shape
```

```python
y = ['sickadvice14']
x = ['education',
#      'emp3',
#      'imputed_povertygroup',
     'insuredgateway14',
     'maritalstatus14',
     'usborn',
     'sexualid',
     #'nutrition1',
     #'nsodaperday14',
#      'pcptalkeathabit14',
#      'activelevel14',
     'athomelanguage14',
    ]
# continuous_x = ['nutrition1','nsodaperday14']
df2 = df.ix[:, y+x]
#Remove observations where independent variables are NAN.
df2 = df2.dropna()
#dummify categorical variables
df2 = pd.get_dummies(df2,columns=x)
```

```python
print "Shape of Data: ",df2.shape
df_x = df2.ix[:, df2.columns != y[0]]
df_y= df2.ix[:,y]
X = df_x.as_matrix()
Y = df_y.as_matrix()

# --------------------
# Normalizing data
#b --------------------
# Xtmax = np.amax(X,axis=0)
# X = np.divide(X,Xtmax)
# --------------------
# K-fold CV
# --------------------
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7, random_state=3)
ts = Y_test.shape[0]

# --------------------
# Naive Bayes
# --------------------
gnb = naive_bayes.BernoulliNB()
gnb.fit(X_train,Y_train)
ypred_gnb = gnb.predict(X_test)
e_gnb = np.sum((ypred_gnb[i] != Y_test[i]) for i in range(0,ts))
print("----------Naive Bayes----------")
print e_gnb, "misclassified data out of",ts,"(",float(e_gnb)*100/ts,"%)"
```

```python
# --------------------
# SVM
# --------------------
svm_linear = svm.SVC(kernel='linear')
svm_linear.fit(X_train,Y_train)
ypred_svm_linear = svm_linear.predict(X_test)
e_svm_linear = np.sum((ypred_svm_linear[i] != Y_test[i]) for i in range(0,ts))
print("----------SVM Linear----------")
print e_svm_linear, "misclassified data out of", ts, "({:0.2f}".format(float(e_svm_linear)*100/ts),"%)"
```

```python
class_weight={1:10} #give a larger importance to class
svm_rbf = svm.SVC(kernel='rbf', gamma=500, class_weight={1:10})
# svm_rbf = svm.SVC(kernel='rbf', gamma=500)
svm_rbf.fit(X_train,Y_train)
ypred_svm_rbf = svm_rbf.predict(X_test)
e_svm_rbf = np.sum((ypred_svm_rbf[i] != Y_test[i]) for i in range(0,ts))
print("----------SVM RBF----------")
print("number of support vectors",len(svm_rbf.support_))
print e_svm_rbf, "misclassified data out of", ts, "(",float(e_svm_rbf)*100/ts,"%)"
```

```python
plt.subplot(1,3,1)
plt.scatter(Y_test,ypred_gnb,c=ypred_gnb)
plt.subplot(1,3,2)
plt.scatter(Y_test,ypred_svm_linear,c=ypred_svm_linear)
plt.subplot(1,3,3)
plt.scatter(Y_test,ypred_svm_rbf,c=ypred_svm_rbf)
# svm_rbf_sv = svm_rbf.support_
# plt.scatter(X[svm_rbf_sv,2],X[svm_rbf_sv,3],color="red")
plt.show()
```

```python
# --------------------
# K-fold CV
# --------------------
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7, random_state=3)
ts = Y_test.shape[0]

# --------------------
# Naive Bayes
# --------------------
gnb = naive_bayes.BernoulliNB()
gnb.fit(X_train,Y_train)
ypred_gnb = gnb.predict(X_test)
e_gnb = np.sum((ypred_gnb[i] != Y_test[i]) for i in range(0,ts))
print("----------Naive Bayes----------")
print e_gnb, "misclassified data out of",ts,"(",float(e_gnb)*100/ts,"%)"
```

```python
IS_incorrect=[]
OS_incorrect=[]
ypreds = []

for count in range(10):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = .05)
    ts_test = Y_test.shape[0]
    ts_train = Y_train.shape[0]
    #     print ts
    #     print X_train.shape[0]

    gnb = naive_bayes.BernoulliNB()
    gnb.fit(X_train,Y_train)
    ypred_gnb = gnb.predict(X_test)
    os_incorrect = np.sum((ypred_gnb[i] != Y_test[i]) for i in range(0,ts_test))
    percent_incorrect_OS = float(os_incorrect)*100/ts_test
    ypred_ISnb = gnb.predict(X_train)
    ypreds.append(ypred_ISnb)
    is_incorrect = np.sum((ypred_ISnb[i] != Y_train[i]) for i in range(0,ts_train))
    percent_incorrect_IS = float(is_incorrect)*100/ts_train
    OS_incorrect.append(percent_incorrect_OS)
    IS_incorrect.append(percent_incorrect_IS)
```

```python
n = 2
pca = PCA(n)
Xproj = pca.fit_transform(X_train)
x, y = Xproj[:,0],Xproj[:,1]
labels = ypred_gnb
df = pd.DataFrame(dict(x=x, y=y, label=labels))

groups = df.groupby('label')

# Plot
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(10)
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=5, label=sick_advice_dict.get(name)[0])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_title('Naive Bayes Classification on 80% Training Set Visualized by Principal Components')
```
